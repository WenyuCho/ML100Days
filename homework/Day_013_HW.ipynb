```python
import pandas as pd
score_df = pd.DataFrame([[1,56,66,70], [2,90,45,34], [3,45,32,55], [4,70,77,89], [5,56,80,70],
                         [6,60,54,55], [7,45,70,79], [8,34,77,76], [9,25,87,60], [10,88,40,43]],
                        columns=['student_id','math_score','english_score','chinese_score'])
score_df
```

```python
#1.6號學生(student_id=6)3科平均分數為何? 56.333333
score_df = score_df.set_index('student_id')
ms = score_df.mean(axis=1)
print("6號學生3科平均分數:" + str(ms[6]) )
```

```python
#2.6號學生3科平均分數是否有贏過班上一半的同學? 否
median_score = score_df.mean(axis=1)
median_score.median()
if ms[6] >= median_score.median():
    print("yes")
else:
    print("no")
```

```python
#3.由於班上同學成績不好，所以學校統一加分，加分方式為開根號乘以十，請問6號同學3科成績分別是? 
#math:77.459667, english:73.484692, chinese:74.161985
new_score = score_df.apply(lambda x : x**(0.5)*10)
new_score
```

```python
#4.承上題，加分後各科班平均變多少?
new_score.mean()
```

- 依照 fruit 欄位做合併
- 依照 index 索引做合併

```python
import numpy as np
import pandas as pd
df1 = pd.DataFrame({
    'fruit': ['apple', 'banana', 'orange'] * 3,
    'weight': ['high', 'medium', 'low'] * 3,
    'price': np.random.randint(0, 15, 9)
})

df2 = pd.DataFrame({
    'fruit': ['apple', 'orange', 'pine'] * 2,
    'weight': ['high', 'low'] * 3,
    'price': np.random.randint(0, 15, 6)
})
print(df1)
print(df2)
```

```python
# - 依照 fruit 欄位做合併
pd.merge(df1, df2, on='fruit')
```

```python
# - 依照 index 索引做合併
df1.join(df2)
```

# 2. [簡答題] 承上題，請問為什麼依照 merge 合併後有些資料不見了？
因為marge的資料只會存取兩邊同時有的資料，如果資料只存在單一邊就不會合併

# 3. [簡答題] 承上題，請問為什麼依照 index 合併會發生錯誤？請用程式解決。
因為join時若有重複的欄位在兩個部分，就無法合併，此時要把左、右相同的欄位命名來排除相同的欄位

```python
df1.join(df2, lsuffix='_df1', rsuffix='_df2')
```
